```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
numpyDizisi2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, color = "black", alpha = 0.3)
benimEksen.plot(numpyDizisi2, numpyDizisi1, color = "blue", alpha = 0.8)
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 1, color = "blue", linestyle = "-.", linewidth = 3)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color = "yellow",linestyle = "-", linewidth = 3)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color = "blue", linestyle = ":", linewidth = 3)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "yellow", linestyle = "--", linewidth = 3)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "blue", linestyle = "--", linewidth = 3, marker = "o", markersize = 8, markerfacecolor = "yellow")
```
