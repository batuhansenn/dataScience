```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]
```

```python
weightList = [85, 90, 72, 81, 84, 82, 90, 95, 72, 68]
```

```python
numpyAge = np.array(ageList)
numpyWeight = np.array(weightList) 
```

```python
numpyAge
```

```python
numpyWeight
```

```python
plt.plot(numpyAge, numpyWeight, 'g')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight')
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, 'g+-')
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1, numpyDizisi2, 'r*-')
plt.subplot(1,2,2)  
plt.plot(numpyDizisi2, numpyDizisi1, 'g--')
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2, 0.2, 0.9, 0.9])
figureAxes.plot(numpyDizisi1, numpyDizisi2, 'g')
figureAxes.set_xlabel('X Axis')
figureAxes.set_ylabel('Y Axis')
figureAxes.set_title('Title')
```

```python
figur2 = plt.figure()

eksen1 = figur2.add_axes([0.1, 0.1, 0.9, 0.9])
eksen2 = figur2.add_axes([0.2, 0.5, 0.3, 0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, 'g')
eksen1.set_xlabel('X Ekseni')
eksen1.set_ylabel('Y Ekseni')
eksen1.set_title('Ana Grafik Başlık')

eksen2.plot(numpyDizisi2, numpyDizisi1, 'b')
eksen2.set_xlabel('X Ekseni')
eksen2.set_ylabel('Y Ekseni')
eksen2.set_title('Küçük Grafik Başlık')
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1, numpyDizisi2, 'g')
    eksen.set_xlabel('X Ekseni')
    
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1, 0.1, 0.9, 0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, label = 'numpyDizisi ** 2')
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label = 'numpyDizisi ** 3')
yeniEksen.legend(loc = 2)
```

```python
yeniFigur.savefig('benimfigur.png', dpi = 200)
```
